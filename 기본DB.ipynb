```python
# 기본
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from matplotlib import font_manager, rc
import platform
from tqdm import tqdm

# crawling
import requests
import lxml.html
import sqlite3
from pandas.io import sql
from bs4 import BeautifulSoup
```

```python
fourth = pd.read_csv('기본DB_4기 (2007-2009).csv', low_memory=False)
fourth
```

```python
fourth_08 = fourth.loc[fourth['year'] == 2008].reset_index(drop=True)
fourth_08
```

```python
fourth_09 = fourth.loc[fourth['year'] == 2009].reset_index(drop=True)
fourth_09
```

```python
data_0809 = pd.concat([fourth_08, fourth_09], axis = 0)
data_0809
```

```python
fifth = pd.read_csv('기본DB_5기 (2010-2012).csv', low_memory=False)
fifth
```

```python
fifth_10 = fifth.loc[fifth['year'] == 2010].reset_index(drop=True)
fifth_10
```

```python
fifth_11 = fifth.loc[fifth['year'] == 2011].reset_index(drop=True)
fifth_11
```

```python
basicDB = pd.concat([data_0809, fifth_10, fifth_11], axis = 0)
basicDB
```

```python
basicDB.to_csv('기본DB (2008-2011).csv', encoding = 'euc-kr', index = False)
```
